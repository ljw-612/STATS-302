```python
##import necessary libraries
import pandas as pd
import numpy as np
import os
import glob
import xml.etree.ElementTree as ET 
```

```python
##direct the colab file to the google drive
from google.colab import drive
drive.mount('/content/drive')
```

```python
#path we store all the annotations
path_an = "/content/drive/MyDrive/STATS302_FINAL_PROJECT/annotations"
```

```python
##extract label message from the annotations
dataset = {
            "xmin":[],
            "ymin":[],   
            "xmax":[],
            "ymax":[],
            "name":[],    
            "file":[],
            "width":[],
            "height":[],
           }

for anno in glob.glob(path_an+"/*.xml"):
    tree = ET.parse(anno)
    
    for elem in tree.iter():
        if 'size' in elem.tag:
            for attr in list(elem):
                if 'width' in attr.tag: 
                    width = int(round(float(attr.text)))
                if 'height' in attr.tag:
                    height = int(round(float(attr.text)))    

        if 'object' in elem.tag:
            for attr in list(elem):
                
                if 'name' in attr.tag:
                    name = attr.text                 
                    dataset['name']+=[name] #label of each face
                    dataset['width']+=[width] #width of each face
                    dataset['height']+=[height] #height of each face
                    dataset['file']+=[anno.split('/')[-1][0:-4]]  #file name of each face
                            
                if 'bndbox' in attr.tag:
                    for dim in list(attr):
                        if 'xmin' in dim.tag:
                            xmin = int(round(float(dim.text)))
                            dataset['xmin']+=[xmin] #left bound of the face
                        if 'ymin' in dim.tag:
                            ymin = int(round(float(dim.text)))
                            dataset['ymin']+=[ymin] #lower bound of the face                               
                        if 'xmax' in dim.tag:
                            xmax = int(round(float(dim.text)))
                            dataset['xmax']+=[xmax] #right bound of the face                              
                        if 'ymax' in dim.tag:
                            ymax = int(round(float(dim.text)))
                            dataset['ymax']+=[ymax] #upper bound of the face   
```

```python
data0=pd.DataFrame(dataset)
data0
```
